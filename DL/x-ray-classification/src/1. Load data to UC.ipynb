Dataset: https://www.kaggle.com/datasets/paultimothymooney/chest-xray-pneumonia

```python
CATALOG = dbutils.widgets.get("CATALOG")
SCHEMA = dbutils.widgets.get("SCHEMA")
VOLUME = dbutils.widgets.get("VOLUME")
```

```python
import subprocess
import zipfile
import os
import logging

logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
)

zip_path = f"/Volumes/{CATALOG}/{SCHEMA}/{VOLUME}/chest-xray-pneumonia.zip"
out_dir = f"/Volumes/{CATALOG}/{SCHEMA}/{VOLUME}/chest-xray-pneumonia"

logging.info("Running: downloading dataset with curl")

subprocess.run(
    [
        "curl",
        "-L",
        "https://www.kaggle.com/api/v1/datasets/download/paultimothymooney/chest-xray-pneumonia",
        "--output",
        zip_path,
    ],
    check=True,
)

logging.info("Finished: download complete")

logging.info("Running: creating output directory if needed")
os.makedirs(out_dir, exist_ok=True)
logging.info("Finished: directory ready")

logging.info("Running: unzipping dataset")
with zipfile.ZipFile(zip_path, "r") as zf:  # unzip
    zf.extractall(out_dir)
logging.info("Finished: unzip complete")
```
