```python
# Install dependencies as needed:
# pip install kagglehub[pandas-datasets]
import kagglehub
from kagglehub import KaggleDatasetAdapter

# Set the path to the file you'd like to load
file_path = "Scrapped_data.csv"

# Load the latest version
df = kagglehub.load_dataset(
  KaggleDatasetAdapter.PANDAS,
  "bishop36/jobs-data",
  file_path,
  # Provide any additional arguments like 
  # sql_query or pandas_kwargs. See the 
  # documenation for more information:
  # https://github.com/Kaggle/kagglehub/blob/main/README.md#kaggledatasetadapterpandas
)

print("First 5 records:", df.head())
```

```python
spark_df = spark.createDataFrame(df)
display(spark_df)
```

```python
from pyspark.sql.functions import concat_ws

spark_df = spark_df.withColumn("job_id", concat_ws("_", spark_df["Job Name"], spark_df["Company name"]))
display(spark_df)
```

```python
# Rename columns with invalid characters
spark_df = spark_df.withColumnRenamed("Job Name", "Job_Name") \
    .withColumnRenamed("Company name", "Company_Name") \
    .withColumnRenamed("Date Posted", "Date_Posted") \
    .withColumnRenamed("Job Function:", "Job_Function") \
    .withColumnRenamed("Hiring Location:", "Hiring_Location")

spark_df.write.mode("overwrite").saveAsTable("workspace.default.jobs")
```
