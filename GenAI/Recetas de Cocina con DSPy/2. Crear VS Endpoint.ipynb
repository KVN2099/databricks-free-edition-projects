```python
from databricks.vector_search.client import VectorSearchClient

vs_endpoint_name = dbutils.widgets.get("vs_endpoint_name")
vsc = VectorSearchClient(disable_notice=True)
```

```python
def endpoint_exists(vsc, vs_endpoint_name):
  try:
    return vs_endpoint_name in [e['name'] for e in vsc.list_endpoints().get('endpoints', [])]
  except Exception as e:
    #Temp fix for potential REQUEST_LIMIT_EXCEEDED issue
    if "REQUEST_LIMIT_EXCEEDED" in str(e):
      print("WARN: couldn't get endpoint status due to REQUEST_LIMIT_EXCEEDED error. The demo will consider it exists")
      return True
    else:
      raise e
```

```python
if not endpoint_exists(vsc, vs_endpoint_name):
  vsc.create_endpoint_and_wait(vs_endpoint_name, endpoint_type='STANDARD', verbose=True)
else:
  print(f"Endpoint named {vs_endpoint_name} already exists.")

print(f"Endpoint named {vs_endpoint_name} is ready.")
```
